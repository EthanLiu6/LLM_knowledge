## 1. S形激活

### 1.1 Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))
```

```python
x = np.random.randn(1000)
x.min()
```

```python
res = sigmoid(x)
res.sort()
```

```python
plt.plot(res)
plt.title('sigmid with random sorted nums')
plt.show()
```

```python
x = np.arange(-8, 8, 0.2)
# x
```

```python
res = sigmoid(x)
# res.sort()
```

```python
plt.plot(res)
plt.title('sigmid with arange(-8, 8, 0.2) nums')
plt.show()
```

```python
def d_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
d_sgmd = d_sigmoid(x)

plt.plot(d_sgmd)
plt.title('d_sigmoid')
plt.show()
```

### 1.2 Tanh

```python
def tanh(x):
    x_exp = np.exp(x)
    return (x_exp - 1 / x_exp) / (x_exp + 1 / x_exp)
```

```python
tanh_res = tanh(x)
```

```python
plt.plot(tanh_res)
plt.title('Tanh')
plt.show()
```

```python
def d_tanh(x):
    return 1 - (tanh(x)**2)
```

```python
d_tanh_res = d_tanh(x)
```

```python
plt.plot(d_tanh_res)
plt.title('d_Tanh')
plt.show()
```
